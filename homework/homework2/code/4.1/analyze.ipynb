```python
import matplotlib.pyplot as plt
```

```python
# Read the file and extract data
file_path = 'data.txt'  # Update with the correct file path

x_values = []
y_values = []

with open(file_path, 'r') as file:
    for line in file:
        line = line.replace(" - ",",").replace("Time for pingpong: ", "").replace(" (microsec)", "")
        x, y = line.split(",")
        x_values.append(int(x))
        y_values.append(float(y))
```

```python

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='Pingpong Time')
plt.xlabel('Processes')
plt.ylabel('Time (microseconds)')
plt.title('Pingpong Time vs Iteration')
plt.legend()
plt.grid()
plt.show()
```
