```python
import matplotlib.pyplot as plt
```

```python
# Read the file and extract data
file_path1 = 'opt1_output.txt'  # Update with the correct file path
file_path2 = 'opt2_output.txt'  # Update with the correct file path

def extract_values(file_path):
    x_values = []
    y_values = []

    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split('-')
            procno = int(parts[1].split(':')[1])
            time = float(parts[2].replace("Time for pingpong: ", "").replace(" (microsec)", ""))
            x_values.append(procno)
            y_values.append(time)

    # Sort the data by x-axis (procno)
    sorted_data = sorted(zip(x_values, y_values))
    x_values, y_values = zip(*sorted_data)
    average = sum(y_values)/len(y_values)
    return x_values, y_values, average

x_1, y_1, average1 = extract_values(file_path1)
x_2, y_2, average2 = extract_values(file_path2)
```

```python
print(f"Average1: {average1}")
print(f"Average2: {average2}")
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_1, y_1, marker='o', linestyle='-', color='b', label='Pingpong Time - Set 1')

plt.xlabel('Source Processor Number')
plt.ylabel('Time (microseconds)')
plt.title('Pingpong Time vs Processor Number')
plt.ylim([0,1.5])
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_2, y_2, marker='s', linestyle='--', color='r', label='Pingpong Time - Set 2')

plt.xlabel('Source Processor Number')
plt.ylabel('Time (microseconds)')
plt.title('Pingpong Time vs Processor Number')
plt.ylim([0,1.5])
plt.legend()
plt.grid()
plt.show()
```
